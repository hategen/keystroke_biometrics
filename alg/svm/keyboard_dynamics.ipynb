```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
keyboard_data = pd.read_csv('metrics.csv')
```

```python
textarea_data = keyboard_data[(keyboard_data['input']=='hat_area')]
```

```python
keyboard_data['timestamp'] = pd.to_numeric(keyboard_data['timestamp'])
keyboard_data['keyDownTimestamp'] = pd.to_numeric(keyboard_data['keyDownTimestamp'])
keyboard_data['keyUpTimestamp'] = pd.to_numeric(keyboard_data['keyUpTimestamp'])
```

```python
textarea_data.head(20)
```

```python
flightTimes =  textarea_data[['keyDownTimestamp','keyUpTimestamp']].diff()
textarea_data['keyDownFlightTime'] = flightTimes['keyDownTimestamp']
textarea_data['keyUpFlightTime'] = flightTimes['keyUpTimestamp']
```

```python
textarea_data.head(20)
```

```python
hatString = 'they can ask me to get just about any hat they like'
```

```python
def getBigrams(sampleString):
    bigramSet = set()
    for i in range( len(sampleString)-1 ):
        bigramSet.add(sampleString[i]+sampleString[i+1])
    print(bigramSet)

getBigrams(hatString)
```

```python

data = []
for index, row in textarea_data.iterrows():
    if index != 0:
        prevRow = textarea_data.loc[index-1]
        #key = prevRow['key'] + row['key']
        keydownToKeydown = row['keyDownFlightTime']
        keyupToKeyup = row['keyUpFlightTime']
        keydownToKeyup = row['keyUpTimestamp'] - prevRow['keyDownTimestamp']
        keyupToKeyDown = row['keyDownTimestamp'] - prevRow['keyUpTimestamp']
        keypressDuration = row['keypressDuration'] + prevRow['keypressDuration']
        lastKeypressDuration = row['keypressDuration']
       
        result = 0
        if row['login'] == 'andrii':
            result = 1;
            
        metric = [keydownToKeydown,keyupToKeyup,keydownToKeyup,keyupToKeyDown, keypressDuration,result]
        data.append(metric)

            
        #print(data)
        
#print(data)
metricsDataframe = pd.DataFrame(data = data, columns=['keydownToKeydown','keyupToKeyup','keydownToKeyup', 'keyupToKeyDown', 'keypressDuration','result'])
```

```python
metricsDataframe
```

```python
metricsDataframe.head()
```

```python
X = metricsDataframe
y = metricsDataframe['result']
X.drop('result',axis=1)
```

```python
y
```

```python
from  sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state = 111)
```

```python
from sklearn.svm import SVC
```

```python
model =  SVC()
```

```python
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,prediction))
print('\n')
print(classification_report(y_test,prediction))
```

```python
from  sklearn.grid_search import GridSearchCV
```

```python
param_grid = {'C':[0.1,1.0,10,100,1000],'gamma':[1,0.1,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_prediction = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_prediction))
print('\n')
print(classification_report(y_test,grid_prediction))
```
