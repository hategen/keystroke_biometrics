```python

keyboard_map = {' ': {'x':5,'y':4},'q': {'x':0, 'y':0}, 'w': {'x':1, 'y':0}, 'e': {'x':2, 'y':0}, 'r': {'x':3, 'y':0}, 't': {'x':4, 'y':0}, 'y': {'x':5, 'y':0}, 'u': {'x':6, 'y':0}, 'i': {'x':7, 'y':0}, 'o': {'x':8, 'y':0}, 'p': {'x':9, 'y':0}, 'a': {'x':0, 'y':1},'z': {'x':0, 'y':2},'s': {'x':1, 'y':1},'x': {'x':1, 'y':2},'d': {'x':2, 'y':1},'c': {'x':2, 'y':2}, 'f': {'x':3, 'y':1}, 'b': {'x':4, 'y':2}, 'm': {'x':5, 'y':2}, 'j': {'x':6, 'y':1}, 'g': {'x':4, 'y':1}, 'h': {'x':5, 'y':1}, 'j': {'x':6, 'y':1}, 'k': {'x':7, 'y':1}, 'l': {'x':8, 'y':1}, 'v': {'x':3, 'y':2}, 'n': {'x':5, 'y':2}, }

def dist(a,b):
    X = (keyboard_map[a]['x'] - keyboard_map[b]['x'])**2
    Y = (keyboard_map[a]['y'] - keyboard_map[b]['y'])**2
    return sqrt(X+Y)
     
     
#for i in keyboard_cartesian.keys():
#    for j in keyboard_cartesian.keys():
#        distance_from_i_to_j = [(i, j, dist(i, j))]
#        print(distance_from_i_to_j)

data = []
for index, row in textarea_data.iterrows():
    if index != 0:
        prevRow = textarea_data.loc[index-1]
        #key = prevRow['key'] + row['key']
        keydownToKeydown = row['keyDownFlightTime']
        keyupToKeyup = row['keyUpFlightTime']
        keydownToKeyup = row['keyUpTimestamp'] - prevRow['keyDownTimestamp']
        keyupToKeyDown = row['keyDownTimestamp'] - prevRow['keyUpTimestamp']
        keypressDuration = row['keypressDuration'] + prevRow['keypressDuration']
        lastKeypressDuration = row['keypressDuration']
        keyDist = dist(prevRow['key'],row['key'])
        result = 0
        if row['login'] == 'andrii':
            result = 1;
            
        metric = [keyDist,keydownToKeydown,keyupToKeyup,keydownToKeyup,keyupToKeyDown, keypressDuration,result]
        if row['login']==prevRow['login'] and keydownToKeydown<1500:
            data.append(metric)

            
        #print(data)
        
#print(data)
metricsDataframe = pd.DataFrame(data = data, columns=['keyDist', 'keydownToKeydown','keyupToKeyup','keydownToKeyup', 'keyupToKeyDown', 'keypressDuration','result'])
```

```python
X = metricsDataframe
y = metricsDataframe['result']
X.drop('result',axis=1)
```

```python
from  sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state = 111)
```

```python
from sklearn.svm import SVC
```

```python
model =  SVC()
```

```python
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,prediction))
print('\n')
print(classification_report(y_test,prediction))
```

```python
from  sklearn.grid_search import GridSearchCV
```
